```python
from math import sin, cos, sqrt, atan2, radians

def dist(row):
    lat1 = radians(row['ToLat'])
    lon1 = radians(row['ToLong'])
    lat2 = radians(row['FromLat'])
    lon2 = radians(row['FromLong'])
    R = 6373.0
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

def wind_class(speed):
    if(speed < 5):
        return "calm"
    if(speed < 11):
        return "light"
    if(speed < 28):
        return "moderate"
    if(speed < 49):
        return "strong"
    else:
        return "very strong"
    
def temp_class(temp):
    if(temp < -10):
        return "very cold"
    if(temp < 0):
        return "cold"
    if(temp < 10):
        return "mild"
    if(temp < 20):
        return "warm"
    if(temp < 30):
        return "hot"
    else:
        return "very hot"
    
```
